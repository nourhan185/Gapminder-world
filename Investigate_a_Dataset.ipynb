## Exploratory Data Analysis



### Research Question 1 (The impact of the aid received on the GDP in Egypt during the period from 2000 to 2010.)

#### To answer this question i need to customize the data for Egypt in the spacific rane of years 

```python
# query about the recieved aids for EGYPT only
aid_recieved1[aid_recieved1.country=="Egypt"].head()
```

```python
# rename the data for the custom contry
Aid = aid_recieved1[aid_recieved1.country=="Egypt"]
```

```python
#query about the recieved aids for a range of years for Egypt
Aid[['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010']]
```

```python
#rename the new data
Aids = Aid[['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010']]
```

```python
# now I get the customized data i need from aid recieved dataframe
Aids.head()
```

```python
# query about the gdp for EGYPT only
GDP2[GDP2.country=="Egypt"].head()
```

```python
#rename the data
Growth = GDP2[GDP2.country=="Egypt"]
```

```python
#query about the gdp for a range of years for Egypt
Growth[['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010']]
```

```python
#rename the customized data
gdp = Growth[['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010']]
```

```python
gdp.head()
```

```python
Aid_recieved0= Aids
Gdp0 = gdp
plt.scatter(Aid_recieved0,Gdp0, c='blue',linewidth=5, alpha=0.8)
plt.title('the Aidrecieved impact on gdp')
plt.xlabel('Aid_recieved0')
plt.ylabel('Gdp0')
plt.tight_layout()
plt.show()
```

#### Acording to the data shape in visualization , the aid recieved doesn't has impact for gdp 

### Research Question 2  (The highest growth contries in 2010!)

```python
#sort data for top 10 gdp in 2010
def top_gdp():
    top_gdp=GDP2.sort_values('2010', ascending=False).head(10)
    return top_gdp
```

```python
top_gdp()
```

```python
top_gdp().plot.barh(x='country', y='2010',title='top 10 Gdp ', color='red')
plt.show(block=True)
```

### As the bars show the top ten contries in gdp for 2010 are :
Oman-Qatar- India- Turkey- Zembabwe-Turkmenistan-Timor-leste-Congo-Peru and Rwanda
